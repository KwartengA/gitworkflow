# **Data Visualization Techniques**

### Pie Chart Visualization

```python
new_dataset = dataset.dropna()

data = new_dataset['quality'].value_counts()

plt.figure(figsize=(9, 5))
plt.pie(data, labels=data.index, autopct='%1.0f%%', startangle=150)
plt.title('QUALITY DISTRIBUTION OF WINE')
plt.axis('equal')  
plt.show()
```

## _Pie Chart Assessment_
---

 
 *  The red wine quality spectrum is complex and varied, ranging from 5.0 to 8.0. As depicted in the pie chart, a substantial portion, totaling 42% and 40%, falls within this range.
 * 
 Wines rated at the highest quality (7.0) constitute 13% of the distribution, whereas those with the lowest ratings (3.0 and 8.0) make up only 1%
  * 
The chart reveals that the majority of red wines are clustered in the moderate to above-avera y rane.ge.

## Bar Chart Visualization Technique for the Residual Sugar Distribution of Red wine

```python
new_dataset = dataset.dropna()

# Count the occurrences in each category
sugar_distribution = dataset['residual sugar'].value_counts()

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(sugar_distribution.index, sugar_distribution, color=['lightcoral', 'green', 'grey', 'yellowgreen', 'blue'])
ax.set_title('Residual Sugar Distribution')
ax.set_xlabel('Residual Sugar (mg/mL)')
ax.set_ylabel('Prevalence')

plt.show()
```

## Assessment of Chart Visualization

-  
The visual representation above illustrates a multitude of entries for red wine residual sugar. Given the complexity of the graph, I have organized the data into more easily discernible categorie
 -  
 'low' for residual sugars ranging from 1.5 to 3.5, 'medium' for those falling between 5.0 and 9.9, and 'high' for residual sugars within the 10.0 to 13.0 range.

```python
# instantiating the ranges for grouping the residual sugar values in the dataset
low_range = (1.5, 3.5)
medium_range = (5.0, 9.9)
high_range = (10.0, 13.0)


# Categorize residual sugar values based on the defined ranges
dataset['sugar_category'] = pd.cut(dataset['residual sugar'], bins=[0] + list(high_range) + [float('inf')], labels=['small', 'medium', 'high'])

plt.figure(figsize=(8, 4))
plt.bar(sugar_distribution.index, sugar_distribution, width=0.6, color='green')
plt.title('Residual Sugar Distribution of Red Wine')
plt.xlabel('Sugar Category')
plt.ylabel('Prevalence')
plt.xticks(range(3), ['small', 'medium', 'high'])
plt.ylim(0, 14)

plt.show()
```

## Analysis Derived
---


- The red wine dataset presents a noteworthy distribution of residual sugar contents, categorized into three classes: small (1.5-3.5), medium (5.0-9.9), and high (10.0-13.0). A considerable portion of the dataset exhibits lower residual sugar content, signifying that the majority of red wines lean towards the low end of the spectrum. Specifically, most red wines showcase low residual sugar levels, with moderate and low representations in the medium and high categories, respectively.

- This distribution insight contributes significantly to our understanding of the dataset, shedding light on prevalent characteristics and preferences associated with red wines. By providing detailed information on the prevalence of distinct residual sugar levels, this analysis enhances our grasp of the common attributes that define red wines and the preferences within this dataset.



## Histogram Visualization technique for Content of Alcohol
<hr>

```python
alcohol_content = new_dataset['alcohol']

plt.figure(figsize=(10, 6))
plt.hist(alcohol_content, bins=15, color='green', edgecolor='black')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.title('Alcohol Content Distribution in Red Wine')

plt.show()
```

## Analysis of Technique
---

- Examining the graph provides a clear indication that a predominant portion of red wines falls within the range of 1.0 to 4.0 in terms of alcohol content. This is followed by instances where the alcohol content is precisely at 10.0, succeeded by values such as 6.0 and 9.0, among others.
  
 - The noteworthy observation emerges as a substantial proportion of red wines display an alcohol content situated between 9.0 and 10.0. This suggests that the majority of red wines in the dataset have an alcohol content falling within the specific bracket of 9.0 to 13.0.
